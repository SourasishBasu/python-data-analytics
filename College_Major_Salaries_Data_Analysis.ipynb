```python
spread = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
clean_df.insert(1, 'Salary Spread', spread)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Salary Spread')
low_risk[['Undergraduate Major', 'Salary Spread']].head()
```

```python
highest_90 = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_90[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head(5) # Highest 90th percentile majors
```

```python
high_risk = clean_df.sort_values('Salary Spread', ascending=False)
high_risk[['Undergraduate Major', 'Salary Spread']].head(10) # Highest spread in salaries
```
