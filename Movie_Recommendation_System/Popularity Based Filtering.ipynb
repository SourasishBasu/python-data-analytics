## Popularity Based Filtering

```
import pandas as pd
movies = pd.read_csv('movies.csv')
```

```
movies
```

## Calculating a Weighted Rating

WR = (v + (v+m)) x R + (m + (v+m)) x C

v - number of votes for a movie

m - minimum number of votes required

R - average rating of the movie

C - average rating across all the movies

```
m = movies['vote_count'].quantile(0.9)
m
```

```
C = movies['vote_average'].mean()
C
```

```
movies_filtered = movies.copy().loc[movies['vote_count'] >= m]
movies_filtered
```

```
def weighted_rating(df, m=m, C=C):
    R = df['vote_average']
    v = df['vote_count']
    wr = ((v / v+m) * R) + (m / (v+m) * C)
    return wr
```

```
movies_filtered['weighted_rating'] = movies_filtered.apply(weighted_rating, axis=1)
movies_filtered
```

```
movies_filtered.sort_values('weighted_rating', ascending=False)[['title', 'weighted_rating']].head(10)
```

```
# Return a dictionary of the top N=num movies

def most_popular(num):
    top = movies_filtered.sort_values('weighted_rating', ascending=False)[['title']].head(num).to_dict()
    return top
```

```
most_popular(15)
```
